Perkembangan Artificial Intelligence (AI) telah memberikan dampak besar dalam berbagai bidang, salah satunya adalah pengenalan ekspresi wajah (Facial Emotion Recognition). Teknologi ini memungkinkan komputer mengenali kondisi emosi seseorang berdasarkan citra wajah menggunakan algoritma Deep Learning.

Pada proyek ini dikembangkan sebuah aplikasi bernama Moodify AI yang mampu mendeteksi ekspresi wajah pengguna secara real-time menggunakan webcam. Setelah emosi dikenali, sistem akan memberikan rekomendasi lagu yang sesuai dengan kondisi emosi pengguna.

Model yang digunakan merupakan Convolutional Neural Network berbasis MobileNetV2 yang dilatih menggunakan dataset FER2013 dari Kaggle.

3. Dataset

Dataset yang digunakan adalah:

FER2013 (Facial Expression Recognition 2013)

Sumber:

https://www.kaggle.com/datasets/msambare/fer2013

Dataset terdiri dari tujuh kelas emosi yaitu:

Angry
Disgust
Fear
Happy
Neutral
Sad
Surprise

Jumlah gambar lebih dari 35.000 citra wajah grayscale.

Library yang Digunakan

```python
import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import GlobalAveragePooling2D
```

5. Persiapan Dataset

Dataset dipisahkan menjadi:

Training
Validation
Test

Seluruh gambar diubah menjadi ukuran: 224 x 224 pixel

6. Data Augmentation

Untuk meningkatkan kemampuan model dilakukan data augmentation menggunakan:

```python
ImageDataGenerator(
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True,
    rescale=1./255
)
```

Pembuatan Model

Model menggunakan transfer learning MobileNetV2.

Arsitektur:

```python
Input Image
↓

MobileNetV2

↓

GlobalAveragePooling

↓

Dropout

↓

Dense

↓

Output (7 Emotion)
```
